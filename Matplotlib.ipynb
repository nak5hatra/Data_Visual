### Matplotlib

```python
# importing the necessary libraries.
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randint
%matplotlib inline
```

```python
# generates a linearly separate distribution between 1 and 10
x = np.linspace(1, 10, 20)
x
```

```python
# generates a random array of 20 elements between 0 and 50
y = randint(1, 50, 20)
y
```

```python
# ploting the y distribution
plt.plot(y)
```

```python
# creating the cumulative sum of y elements
y = y.cumsum()
```

```python
# now ploting the y distribution
plt.plot(y)
```

<i>Note:<i> the `plt.plot()` function usually takes x and y as arguments and plots the result in x-axis and y-axis respectively. However we can also use the `plt.plot()` function to plot the single data points like we did in the previous example: `plt.plot(y)`

```python
# ploting the x-axis and y-axis
plt.plot(x, y)
```

By default plots are plotted with blue color. but if you want to plot in a different color then you can set the color using <i>color='r'<i> here 'r' refers to the red color. we can also see the reach element by using the <i>marker='o'</i> here 'marker' shows the reach element.

```python
plt.plot(x, y, color = 'r', marker='o')
```

```python
# the following plot shows changes in the style of the line by using the linestyle = '--' argument.
plt.plot(x, y, color = 'r', marker='o', linestyle = '--')
```

### Label

Labels are used to give clear understanding of the data we are visually.. the following example shows the labeling of the plot.
`plt.xlabel('your x label')` gives labels for the x axis and `plt.ylabel('your y label')` gives labels for the y axis.
the title of the plot can be changed by setting the `plt.title('Your title')`

```python
plt.plot(x, y, color = 'r', marker='o', linestyle = '--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('This is a title')
plt.show()
```

### Scatter, Bar, Hist, And Box Plots

Here we'll use scatter plot, bar chart, histogram plot and box plot, which will show us the different way to plot the same data. we'll be using same x and y values for this section.

```python
x, y
```

#### Scatter plot
A scatter plot presents the relationship between two variables in a data-set. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

`plt.scatter()` take two arguments x and y. each dot indicates values for an individual data point.

```python
plt.scatter(x, y, color = 'g')
```

#### Bar plot
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

in the example below, list 'b' represents the height of the bar chart and 'a' represents the category of each element of 'b'.
in `plt.bar()` "width" assigns the width of each bar.

```python
b = [10, 23, 43, 5, 66]
a = ['a', 'b', 'c', 'd', 'e']
plt.bar(a,b, width=0.9)
```

#### Histogram 
Histograms are a type of bar plot for numeric data that group the data into bins. in example below, the each bar in the plot shows the bin.

```python
plt.hist(y,rwidth=0.8)
```

#### Box plot

```python
data = [np.random.normal(0, std, 100) for std in range(6, 9)]
# data
```

```python
plt.boxplot(data, vert=True,patch_artist= True)
# plt.boxplot(data[0], vert=True,patch_artist= True)
# plt.boxplot(data[1], vert=True,patch_artist= True)
plt.show()
```

# Subplot
with the help of subplot multiple plots can be created in a single figure. `plt.subplots()` can be used to create multiple plots, subplots takes two arguments. The first one controls the number of rows, the second one the number of columns.

```python
# Using previous data x and y
x,y
```

```python
# creating new data y2 by multiplying y and x
y2 = y*x
y2
```

```python
# ploting y and y2 the top of each other.
plt.plot(y)
plt.plot(y2)
```

```python
plt.subplot(2, 1, 1)
plt.plot(y, color='g', linestyle='--')

plt.subplot(2, 1, 2)
plt.plot(y2, color='r', marker='*', markersize=10)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.7, 0.3, 0.2])

ax1.plot(x, y, color = "r")
ax1.set_xlabel("This is Big X")
ax1.set_ylabel("This is big Y")
ax1.set_title("This is big plot")

ax2.plot(x, y2, color = "g", linestyle = "dashed")
ax2.set_xlabel("This is small X")
ax2.set_ylabel("This is small Y")
ax2.set_title("This is small plot")
plt.show()

```

#### Pie Plot

```python
labels = ["Cat", "Dog", "Birds", "Fish"]
population = [10, 25, 60, 30]
```

```python
fig, ax = plt.subplots( figsize =(3, 3), dpi = 150)
ax.pie(population, labels=labels, autopct="%1.1f%%", explode=(0.08, 0, 0, 0))
plt.show()
```

```python
fig, ax = plt.subplots( figsize =(3, 3), dpi = 150)
patches, texts, autotexts = ax.pie(population, labels=labels, autopct="%1.1f%%", 
       explode=(0.08, 0, 0, 0))
plt.setp(autotexts, size = 8, color = "white")
autotexts[1].set_color("black")
autotexts[1].set_size(15)
plt.show()
```

```python
fig, ax = plt.subplots(dpi= 100)
size = 0.3

values  = np.array([[60, 32], [35,20], [26, 36]])
values_sum = values.sum(axis=1)
values_flat = values.flatten()

color_map = plt.get_cmap('tab20c')
outer_colors = color_map(np.arange(3)*3)
inner_colors = color_map(np.arange(6)*3)
ax.pie(values_sum, radius=1, colors=outer_colors, wedgeprops= dict(width=size, edgecolor='w'))
ax.pie(values_flat, radius=1-size, colors=inner_colors, wedgeprops= dict(width=size, edgecolor='w'))
```

```python
fig, ax = plt.subplots(dpi= 100)
size = 0.3

values  = np.array([[60, 32], [35,20], [26, 36]])
values_sum = values.sum(axis=1)
values_flat = values.flatten()

color_map = plt.get_cmap('tab20c')
outer_colors = color_map(np.arange(3)*3)
inner_colors = color_map(np.arange(6)*3)

ax.pie(values_sum, radius=1, colors=outer_colors, 
       wedgeprops= dict(width=size, edgecolor='w'), autopct="%1.1f%%", pctdistance=0.89)

ax.pie(values_flat, radius=1-size, colors=inner_colors, 
       wedgeprops= dict(width=size, edgecolor='w'), autopct="%1.1f%%", pctdistance=0.69,)
plt.show()
```

```python
fig, ax = plt.subplots(dpi = 100)
recipe = ["350 g Flour", "75 g Suger", "500 g almond", "900 g beef"]

color_map = plt.get_cmap('tab20c')
df = [float(x.split(' ')[0]) for x in recipe]
ingredients = [x.split(' ')[2] for x in recipe]

wedges, texts, autotexts = ax.pie(df, autopct= "%0.1F%%", colors=color_map(np.arange(6)*4))
ax.legend(wedges, ingredients, title = "ingredients", loc = "center left")
```

```python
fig, ax = plt.subplots(dpi = 100)
recipe = ["350 g Flour", "75 g Suger", "500 g almond", "900 g beef"]

color_map = plt.get_cmap('tab20c')
df = [float(x.split(' ')[0]) for x in recipe]
ingredients = [x.split(' ')[2] for x in recipe]

wedges, texts, autotexts = ax.pie(df, autopct= "%0.1F%%", colors=color_map(np.arange(6)*4))
ax.legend(wedges, ingredients, title = "Ingredients", loc = "center left", bbox_to_anchor= (1, 0, 1, 1))
```
