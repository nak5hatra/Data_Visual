### Complete EDA of Titanic  Dataset

The RMS Titanic was a British passenger liner that sank in the North Atlantic Ocean in the early
orning hours Of 15 April 1912. after it collided with an iceberg during its maiden voyage from
thampton to New York City. There were an estimated 2,224 passengers and crew aboard the
ip, and more than I ,500 died.

<li>Name - Name</li>
<li>Sex - Sex</li>
<li>Age - Age</li>
<li>Sibsp - Number of Siblings/Spouses Aboard</li>
<li>Parch - Number of Parents/ChiIdren Aboard</li>
